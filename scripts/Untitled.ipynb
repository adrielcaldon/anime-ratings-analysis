```python
import pandas as pd

df = pd.read_csv("../data/anime_cleaned.csv")
df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
```

```python
top_genres = df['Genres'].value_counts().head(10)

plt.figure(figsize=(10, 6))
sns.barplot(x=top_genres.values, y=top_genres.index, palette="viridis")
plt.title("Top 10 Anime Genres")
plt.xlabel("Number of Titles")
plt.ylabel("Genre")
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Convert to DataFrame so we can use 'hue'
genre_df = pd.DataFrame({
    'Genre': top_genres.index,
    'Count': top_genres.values
})

plt.figure(figsize=(10, 6))
sns.barplot(data=genre_df, x='Count', y='Genre', hue='Genre', palette='viridis', dodge=False)
plt.title("Top 10 Anime Genres")
plt.xlabel("Number of Titles")
plt.ylabel("Genre")
plt.legend([],[], frameon=False)  # Hide legend
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(df['Score'], bins=20, kde=True, color="skyblue")
plt.title("Distribution of Anime Scores")
plt.xlabel("Score")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='Popularity', y='Score', hue='Genres', legend=False, alpha=0.6)
plt.title("Score vs. Popularity")
plt.xlabel("Popularity")
plt.ylabel("Score")
plt.tight_layout()
plt.show()
```
